# pandas- Apply Function

```python
This method allow us to apply function to DF
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randint(1,30,size=(6,4)),columns=list('ABCD'))
df
```

```python
df.apply(lambda x:x**2)
```

```python
df.apply(lambda x: x**3)
```

```python
df.apply(lambda x:x+100)
```

```python
df
```

```python
def cube(x):
    return x**3
df.apply(cube)
```

```python
def cube(x):
    return x+100
df.apply(cube)
```

```python
def squareroot(x):
    return x**0.5
df.apply(squareroot)
```

```python
df.apply(np.sqrt)
```

```python
df.apply(np.cbrt)
```

```python
df
```

```python
df['A']=df['A'].apply(lambda x: x**2)
df
```

```python
df.apply(np.max)  #axis = 0 or 'index' >>> apply to the column 
                 #axis = 1 or 'columns' >>> apply to the row 
```

```python
df.apply(np.max,axis=0)
```

```python
df.apply(np.max,axis='index')
```

```python
df.apply(np.max,axis=1)
```

```python
df.apply(np.min,axis='columns')
```

```python
df.apply(np.min,axis='index')
```

```python
df=pd.DataFrame(np.random.randint(1,30,size=(6,4)),columns=list('ABCD'))

df['A']=df['A'].apply(lambda x:x**2)
df['B']=df['B'].apply(lambda x:x**3)
df['C']=df['C'].apply(lambda x:x+100)
df['D']=df['D'].apply(np.sqrt)
df
```

```python
import re

movie_df=pd.read_csv('movies_data.csv')

def get_movie_titles(x):
    movie_name=re.sub('[^a-zA-z]+',' ',x)
    return movie_name
movies_df['title']=movies_df['title'].apply(get_movie_titles)
movies_df.head(30)
```
