# Linear Regression

```python
Linear line equation
y=mx + c
y= m1x1+ m2x2 + c

y = dependent variable/target var  (contineous >> age,gate score,price)
m = slope 
x= independent variable
c = y intercept
```

```python
It is a predictive model used to find linear relationship between 
dependent variable and one or more independent variable.
- It is a parametric algorithm >> Assumption of distribution of data.
```

```python
Types of linear Regression:
    1. Simple Linear Regression:
        only one independent variable
        y= mx +c
    2. Multiple Linear Regression:
        two or more independent variable
        y = m1x1 + m2x2 +c
```

```python
# Assumptions of Linear Regression:
    1. Linearity
    2. No Multicolinearity
    3. Normality of residual (error)
    4. Homoscedasticity (residual)
```

# 1. Linearity

```python
Linear Relationsip between Independent variable and Dependent Variable
Coefficient of correlation >> Pearson Correlation Coefficient >> R- Value
R value >>> -1 to +1
Corr = Covariance / (product Std Dev)
Rxy = Covariance / (product Std Dev)
Good Predictor = >0.7  or <-0.7
Bad Predictor  = <0.3   or > -0.3
Positive corr >> R is +ve
Negative Corr >> R is -Ve
```
