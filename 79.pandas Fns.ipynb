## iterate dataframe using for loop

```python
array=np.random.randint(20,40,size=(6,4))
df=pd.DataFrame(array,columns=list('abcd'),index=list('python'))
df
```

# 1. df.items()

<!-- >> iterate over colums
>> return column and data content -->

```python
for col_name,col_data in df.items():
    print('Column Name :',col_name)
```

```python
for col_name,col_data in df.items():
    print('Column Name :',col_name)
    print(col_data)
    print('*'*30)
```

# 2. df.iteritems()

```python
for col_name,col_data in df.iteritems():
#     print('Column Name :',col_name)
    print(col_data)
    print('*'*30)
```

```python
for col_name,col_data in df.iteritems():
    print('Column Name :',col_name)
#     print(col_data)
    print('*'*30)
```

```python
for col_name,col_data in df.iteritems():
    print('Column Name :',col_name)
    print(col_data)
    print('*'*30)
```
