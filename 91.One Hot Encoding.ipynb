# Train and Test Data

```python
df1=pd.DataFrame({"CITY":['Pune','Mumbai','Delhi','Kolkata','Mumbai','Delhi','Pune','Mumbai','Delhi','Delhi','Kolkata']})
df1
x_train=df1.head(7)
x_train
```

```python
x_test=df1.iloc[7:]
x_test
```

# Encoding on Training Dataset

```python
OneHot_enc=OneHotEncoder(dtype=int)
x_train_res=OneHot_enc.fit_transform(x_train).toarray()
x_train_df=pd.DataFrame(x_train_res,columns=OneHot_enc.categories_)
x_train_df
```

# Encoding on Test Data

```python
x_test_df=pd.DataFrame(OneHot_enc.transform(x_test).toarray(),columns=OneHot_enc.categories_)
x_test_df
```
