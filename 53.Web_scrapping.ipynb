```python
import re
import pandas as pd
from urllib.request import urlopen
```

```python
url = "https://www.summet.com/dmsi/html/codesamples/addresses.html"
html_string =urlopen(url).read().decode()
print(html_string)
```

```python
#contact number >> (564) 908-6970   pattern '\(\d{3}\)\s\d{3}[-]\d{4}'
# zip code  >> 07528        pattern '\d{5}'

contact_number = re.findall('\(\d{3}\)\s\d{3}[-]\d{4}',html_string)
zipcode = re.findall('\d{5}',html_string)
```

```python
result = {"Contact_Number":contact_number, "ZipCode":zipcode}
df=pd.DataFrame(result)
df
```

```python
df.to_csv('Data.csv',index=False)
```
