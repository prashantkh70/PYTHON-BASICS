## 4. Create Excel file

```python
mov_df=pd.read_csv('Movies_Data.csv')
mov_df
```

```python
movie_rating_df=mov_df[['title','star_rating','content_rating']]
# movie_rating_df
```

```python
movie_genre_df=mov_df[['title','genre','duration']]
# movie_genre_df
```

```python
movie_actors_df=mov_df[['title','actors_list']]
# movie_actors_df
```

```python
movie_rating_df.to_excel('movie_ratings.xlsx',index=False)
```

```python
movie_rating_df.to_excel('movie_ratings.xlsx',sheet_name='rating',index=False)
```

```python
writer=pd.ExcelWriter('Movies_Details.xlsx',engine='xlsxwriter')
movie_rating_df.to_excel(writer,sheet_name='rating',index=False)
movie_genre_df.to_excel(writer,sheet_name='genre',index=False)
movie_actors_df.to_excel(writer,sheet_name='actors',index=False)
writer.save()
```
