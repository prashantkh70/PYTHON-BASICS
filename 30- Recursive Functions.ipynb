### factorial

```python
num = 5
fact =1
for i in range(1,num+1):
    fact*=i
print(f'factorial of {num} is  {fact}')
```

```python
def factorial(n):   #n=5
    if n==1:
        return 1
    else:
        return n * factorial(n-1)    #5 *4*3*2*1
n=5
f=factorial(n)
print(f'factorial of {n} is  {f}')
```

### Fibonacci

```python
0,1,1,2,3,5,8,13,21,......
```

```python
first = 0
second =1
n=5
print('Fibonacci series is: ',first,second,end =' ')
for i in range(n-2):
    first,second=second,first+second
    print(second,end= ' ')
```

```python
first = 0
second =1
n=10
print('Fibonacci series is: ',first,second,end =' ')
for i in range(n-2):
    first,second=second,first+second
    print(second,end= ' ')
```

```python
def fibonacci(n):
    first = 0
    second =1

    print('Fibonacci series is: ',first,second,end =' ')
    for i in range(n-2):
        first,second=second,first+second
        print(second,end= ' ')
    print()

fibonacci(8)
fibonacci(10)
fibonacci(13)
```
